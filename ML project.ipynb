<b>Normalization of BH data</b>

-> Normalise the data  with MinMax Scaler

```python
#prepare the data for linear regression algorithm
#select the columns to use for x and y respectively
X = df['LSTAT'].values
Y = df['target'].values
```

```python
#before normalization
print(Y[:5])
```

```python
#performing normalization
x_scaler = MinMaxScaler()
X = x_scaler.fit_transform(X.reshape(-1,1))
X = X[:, -1]
y_scaler = MinMaxScaler()
Y = y_scaler.fit_transform(Y.reshape(-1,1))
Y = Y[:, -1]
```

```python
#after normalization
print(Y[:5])
```

<b>STEP2: split the data</b>

-> split the data into training and testing data -> cross validation

```python
#splitting and sepeartion of testing dataset
#0.2 indicates 20% of the data is randomly sampled as testing data 
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size = 0.2)
```

<b>STEP3: Define Error Function</b>

<b>Gradient Descent in ML</b> has 3 functions-
1. Update Function
2. Error Function
3. Gradient Descent Function

<b>Update Function</b>

```python
#Define Update Function
def update (m, x, c, t, learning_rate):
    grad_m = sum(2 * ((m * x + c) - t) * x)
    grad_c = sum(2 * ((m * x + c) - t))
    m = m - grad_m * learning_rate
    c = c - grad_c * learning_rate
    return m, c
```

<b>Error function</b>

```python
#define Error function
def error (m, x, c, t):
    N = x.size
    e = sum(((m * x + c) - t) ** 2)
    return e * 1 / (2 * N)
```

<b>Gradient Descent function</b>

```python
#Define Gradient Descent function
def gradient_descent(init_m, init_c, x, t, learning_rate, iterations, error_threshold):
    m = init_m
    c = init_c
    error_values = list()
    mc_values = list()

    for i in range(iterations):
        e = error(m, x, c, t)
        if e < error_threshold:
            print('Error threshold less than threshold. Stop the gradient descent.')
            break
        error_values.append(e)
        m, c = update(m, x, c, t, learning_rate)
        mc_values.append((m,c))
    return m, c, error_values, mc_values
```

<b>Running Gradient Descent function</b>

```python
%%time
init_m = 0.9
init_c = 0
learning_rate = 0.001
iterations = 250
error_threshold = 0.001

m, c, error_values, mc_values = gradient_descent(init_m, init_c, xtrain, ytrain, learning_rate, iterations, error_threshold)
```

<b>STEP4: Model Training Visualization</b>

```python
#As the number of iterations increases, changes in the line are less noticable
#In order to reduce the processing time for the animation, it is advised to choose values
mc_values_anim = mc_values[0 : 250 : 5]
```

<b>Initialize the variables</b>

```python
#initialize figures and axis object
fig, ax = plt.subplots()

#initialize line object ( the same object is used by init and update function)
ln, = plt.plot([], [], 'ro-', animated = True)

#define the init function
def init():
    plt.scatter(xtrain, ytrain, color = 'g')
    ax.set_xlim(0, 1.0)
    ax.set_ylim(0, 1.0)
    return ln,

#define the update function
def update_frame(frame):
    m, c = mc_values_anim[frame]
    x1, y1 = -0.5, m * -.5 + c
    x2, y2 = 1.5, m * 1.5 + c
    ln.set_data([x1,x2], [y1,y2])
    return ln,

#Declare the animated object
#Set the callbacks and number of frames
anim = FuncAnimation(fig, update_frame, frames = range(len(mc_values_anim)), init_func = init, blit = True)

#Generate the video of the animation 
HTML(anim.to_html5_video())
```

<b>STEP5: Prediction</b>

```python
#Calculate the predicted value on the test set as a vectorized operation
predicted  = (m * xtest) + c
```

```python
#Compute MSE for the predicted values on the training set
mean_squared_error(ytest, predicted)
```

```python
#Put xtest, ytest and predicted values into a single DataFrame 
#so that we can see the predicted values alongside the testing data
p = pd.DataFrame(list(zip(xtest, ytest, predicted)), columns = ['x', 'target_y', 'predicted_y'])
p.head()
```

->Revert normalisation to obtain the predicted price of house in 1000 dollars. 
The predicted value are in range of 0-1. This is not very useful to us when we want to obtain the price. 
Use inverse transform to scale the values back to the original representation.

```python
#Reshape  To shape  that is required by the scalar i.e. in vertical orientation(column) instead of horizontal orientation(row)
xtest = xtest.reshape(-1, 1)
ytest = ytest.reshape(-1, 1)
predicted = predicted.reshape(-1, 1)

#Perform inverse transformation function 
xtest_scaled = x_scaler.inverse_transform(xtest)
ytest_scaled = y_scaler.inverse_transform(ytest)
predicted_scaled = y_scaler.inverse_transform(predicted)

#This is to remove the extra dimension 
xtest_scaled = xtest_scaled[:, -1]
ytest_scaled = ytest_scaled[:, -1]
predicted_scaled = predicted_scaled[:, -1]

p = pd.DataFrame(list(zip(xtest_scaled, ytest_scaled, predicted_scaled)), columns = ['x', 'target_y', 'predicted_y'])
p = p.round(decimals = 2)
p.head()
```
